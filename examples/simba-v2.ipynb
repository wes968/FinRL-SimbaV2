```python
# FinRL + SimbaV2 + A2C + PPO ‚Äî ÂñÆËÇ°ÁØÑ‰æã (2330.TW)
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import itertools
import matplotlib.pyplot as plt # Matplotlib inline is for notebooks, so we just import the library

from finrl.meta.preprocessor.yahoodownloader import YahooDownloader
from finrl.meta.preprocessor.preprocessors import FeatureEngineer, data_split
from finrl.meta.env_stock_trading.env_stocktrading import StockTradingEnv
from finrl.config import INDICATORS

# SB3 ÁÆóÊ≥ï
from stable_baselines3 import A2C, PPO

# ---------------------------
# 1. ÂèÉÊï∏Ë®≠ÂÆö & Ë≥áÊñô‰∏ãËºâ
# ---------------------------
TRAIN_START_DATE = '2023-01-01'
TRAIN_END_DATE   = '2024-07-01'
TRADE_START_DATE = '2024-07-01'
TRADE_END_DATE   = '2025-06-08'
# Changed TICKERS to only '2330.TW'
TICKERS = ['2330.TW']

# Download historical data for the specified tickers
df_raw = YahooDownloader(
    start_date  = TRAIN_START_DATE,
    end_date    = TRADE_END_DATE,
    ticker_list = TICKERS
).fetch_data()

# ---------------------------
# 2. ÁâπÂæµÂ∑•Á®ã
# ---------------------------
# Initialize FeatureEngineer with technical indicators and turbulence
fe = FeatureEngineer(
    use_technical_indicator = True,
    tech_indicator_list     = INDICATORS,
    use_vix                 = True,
    use_turbulence          = True,
    user_defined_feature    = False
)
# Preprocess the raw data
processed = fe.preprocess_data(df_raw)
processed['date'] = pd.to_datetime(processed['date'])

# Complete the date x ticker grid to handle missing data by filling with 0
dates = processed['date'].dt.normalize().unique().tolist()
tics  = processed['tic'].unique().tolist()
mesh  = pd.DataFrame(itertools.product(dates, tics), columns=['date','tic'])
processed_full = (
    mesh
    .merge(processed, on=['date','tic'], how='left')
    .sort_values(['date','tic'])
    .fillna(0)
)

# ---------------------------
# 3. ÂàáÂàÜË≥áÊñô
# ---------------------------
# Split data into training and trading sets
train = data_split(processed_full, TRAIN_START_DATE, TRAIN_END_DATE)
trade = data_split(processed_full, TRADE_START_DATE, TRADE_END_DATE)

# ---------------------------
# 4. Âª∫Á´ã Gym Áí∞Â¢É
# ---------------------------
# Determine stock dimensions and state space for the trading environment
stock_dim   = len(train.tic.unique())
state_space = 1 + 2*stock_dim + len(INDICATORS)*stock_dim

# Environment keyword arguments for the StockTradingEnv
env_kwargs = {
    'df': train,
    'hmax': 1000, # Maximum number of shares to buy/sell
    'initial_amount': 1_000_000, # Initial portfolio value
    'buy_cost_pct':    [0.001425] * stock_dim, # Buy cost percentage
    'sell_cost_pct':   [0.001425] * stock_dim, # Sell cost percentage
    'num_stock_shares':[0] * stock_dim, # Initial stock shares
    'state_space':     state_space, # Total state space dimensions
    'stock_dim':       stock_dim, # Number of unique stocks
    'tech_indicator_list': INDICATORS, # List of technical indicators used
    'action_space':    stock_dim, # Action space dimensions (number of stocks)
    'reward_scaling':  1e-4, # Reward scaling factor
    'mode': 'train' # Set mode to train for the training environment
}
e_train_gym = StockTradingEnv(**env_kwargs)

# ---------------------------
# 5. ÂÆöÁæ© SimbaV2 ÊºîÁÆóÊ≥ï
# ---------------------------
# Actor neural network for SimbaV2
class Actor(nn.Module):
    def __init__(self, s_dim, a_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(s_dim, 256), nn.ReLU(),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, a_dim), nn.Softmax(dim=-1) # Softmax for probability distribution over actions
        )
    def forward(self, x):
        return self.net(x)

# Critic neural network for SimbaV2
class Critic(nn.Module):
    def __init__(self, s_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(s_dim, 256), nn.ReLU(),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, 1) # Output a single value (V-value)
        )
    def forward(self, x):
        return self.net(x)

# SimbaV2 Reinforcement Learning algorithm
class SimbaV2:
    def __init__(self, env, gamma=0.99, actor_lr=1e-4, critic_lr=1e-3):
        self.env, self.gamma = env, gamma
        s_dim = env.observation_space.shape[0] # State dimension
        a_dim = env.action_space.shape[0] # Action dimension
        self.actor  = Actor(s_dim, a_dim) # Actor network
        self.critic = Critic(s_dim) # Critic network
        self.opt_a = optim.Adam(self.actor.parameters(), lr=actor_lr) # Optimizer for actor
        self.opt_c = optim.Adam(self.critic.parameters(), lr=critic_lr) # Optimizer for critic

    # Select an action based on the current state
    def select_action(self, s):
        with torch.no_grad():
            return self.actor(torch.FloatTensor(s).unsqueeze(0)).numpy()[0]

    # Train the SimbaV2 model for a given number of steps
    def learn(self, steps):
        obs, _ = self.env.reset() # Reset environment
        for _ in range(steps):
            action = self.select_action(obs) # Select action
            next_obs, reward, term, trunc, _ = self.env.step(action) # Take step in environment
            done = term or trunc # Check if episode is done

            s_t  = torch.FloatTensor(obs).unsqueeze(0) # Current state tensor
            s2_t = torch.FloatTensor(next_obs).unsqueeze(0) # Next state tensor
            r_t  = torch.tensor(reward, dtype=torch.float32) # Reward tensor
            # Calculate target value for critic using Bellman equation
            target = r_t + self.gamma * self.critic(s2_t).detach() * (1-int(done))

            # Critic Update: Minimize MSE loss between predicted value and target
            loss_c = nn.MSELoss()(self.critic(s_t), target)
            self.opt_c.zero_grad(); loss_c.backward(); self.opt_c.step()
            # Actor Update: Maximize expected return ( approximated by critic output * log of policy)
            loss_a = -(self.critic(s_t) * torch.log(self.actor(s_t)+1e-8)).mean()
            self.opt_a.zero_grad(); loss_a.backward(); self.opt_a.step()

            obs = next_obs if not done else self.env.reset()[0] # Update observation
        return self

    # Predict account value during trading phase
    def predict(self, env):
        env.reset() # Reset trading environment
        done = False
        while not done:
            s = env.state # Get current state
            a = self.select_action(s) # Select action
            _, _, term, trunc, _ = env.step(a) # Take step
            done = term or trunc # Check if episode is done
        return env.asset_memory  # Return recorded account values

# ---------------------------
# 6. ÈÄöÁî® DRL Agent
# ---------------------------
class DRLAgent:
    def __init__(self, env):
        self.env = env # Initialize with the Gym environment

    # Get the appropriate RL model based on name
    def get_model(self, name, **kwargs):
        if name == "simbav2":
            return SimbaV2(self.env, **kwargs)
        elif name == "a2c":
            return A2C("MlpPolicy", self.env, verbose=0, **kwargs)
        elif name == "ppo":
            return PPO("MlpPolicy", self.env, verbose=0, **kwargs)
        else:
            raise ValueError(f"Unknown model {name}")

    # Train the selected model
    def train_model(self, model, total_timesteps, name):
        if name == "simbav2":
            return model.learn(total_timesteps) # SimbaV2 has its own learn method
        else:
            model.learn(total_timesteps=total_timesteps) # Stable Baselines3 models
            return model

    @staticmethod
    # Predict account value for a given trained model and trading environment
    def predict_account(model, name, env_kwargs, trade_df):
        args = env_kwargs.copy(); args.update({'df':trade_df, 'mode':'trade'})
        e_trade = StockTradingEnv(**args) # Create a new environment for trading mode

        if name == "simbav2":
            vals = model.predict(e_trade) # SimbaV2's own predict method
        else:
            obs, _ = e_trade.reset() # Reset environment
            done = False
            vals = [e_trade.asset_memory[0]] # Initialize with initial amount
            while not done:
                action, _ = model.predict(obs) # Predict action
                obs, _, term, trunc, _ = e_trade.step(action) # Take step
                done = term or trunc # Check if episode is done
                vals.append(e_trade.asset_memory[-1]) # Append current asset value

        dates = sorted(trade_df['date'].unique())
        # If there's an extra initial value, slice it off
        if len(vals) == len(dates) + 1:
            vals = vals[1:]
        assert len(vals) == len(dates) # Ensure lengths match
        return pd.DataFrame({'date': dates, 'account_value': vals})

# ---------------------------
# 7. ÔºàÈáçË∑ëÔºâËÆ≠ÁªÉ & ÂõûÊµãÂÖ®ÈÉ®ÁÆóÊ≥ï
# ---------------------------
agent = DRLAgent(e_train_gym) # Initialize DRLAgent
algos = {"SimbaV2":"simbav2", "A2C":"a2c", "PPO":"ppo"} # Algorithms to compare
timesteps = 30 # Total timesteps for training
results = {} # Dictionary to store results
for label, name in algos.items():
    print(f"Training and evaluating {label}...")
    model   = agent.get_model(name) # Get model instance
    trained = agent.train_model(model, timesteps, name) # Train model
    df_val  = DRLAgent.predict_account(trained, name, env_kwargs, trade) # Predict account value
    results[label] = df_val # Store results

# ---------------------------
# 8. ‰∏ÄÂº†ÂõæÊØîËæÉ
# ---------------------------
plt.figure(figsize=(10,5)) # Create a figure
for label, df_val in results.items():
    plt.plot(df_val['date'], df_val['account_value'], label=label) # Plot account value for each algorithm
plt.xlabel('Date') # X-axis label
plt.ylabel('Account Value') # Y-axis label
plt.title('Comparison of Different Strategies on TSMC') # Plot title
plt.legend() # Show legend
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show() # Display the plot
```

```python
# FinRL + SimbaV2 + A2C + PPO ‚Äî ÂñÆËÇ°ÁØÑ‰æã (2382.TW)
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import itertools
import matplotlib.pyplot as plt # Matplotlib inline is for notebooks, so we just import the library

from finrl.meta.preprocessor.yahoodownloader import YahooDownloader
from finrl.meta.preprocessor.preprocessors import FeatureEngineer, data_split
from finrl.meta.env_stock_trading.env_stocktrading import StockTradingEnv
from finrl.config import INDICATORS

# SB3 ÁÆóÊ≥ï
from stable_baselines3 import A2C, PPO

# ---------------------------
# 1. ÂèÉÊï∏Ë®≠ÂÆö & Ë≥áÊñô‰∏ãËºâ
# ---------------------------
TRAIN_START_DATE = '2023-01-01'
TRAIN_END_DATE   = '2024-07-01'
TRADE_START_DATE = '2024-07-01'
TRADE_END_DATE   = '2025-06-08'
# Changed TICKERS to only '2382.TW'
TICKERS = ['2382.TW']

# Download historical data for the specified tickers
df_raw = YahooDownloader(
    start_date  = TRAIN_START_DATE,
    end_date    = TRADE_END_DATE,
    ticker_list = TICKERS
).fetch_data()

# ---------------------------
# 2. ÁâπÂæµÂ∑•Á®ã
# ---------------------------
# Initialize FeatureEngineer with technical indicators and turbulence
fe = FeatureEngineer(
    use_technical_indicator = True,
    tech_indicator_list     = INDICATORS,
    use_vix                 = True,
    use_turbulence          = True,
    user_defined_feature    = False
)
# Preprocess the raw data
processed = fe.preprocess_data(df_raw)
processed['date'] = pd.to_datetime(processed['date'])

# Complete the date x ticker grid to handle missing data by filling with 0
dates = processed['date'].dt.normalize().unique().tolist()
tics  = processed['tic'].unique().tolist()
mesh  = pd.DataFrame(itertools.product(dates, tics), columns=['date','tic'])
processed_full = (
    mesh
    .merge(processed, on=['date','tic'], how='left')
    .sort_values(['date','tic'])
    .fillna(0)
)

# ---------------------------
# 3. ÂàáÂàÜË≥áÊñô
# ---------------------------
# Split data into training and trading sets
train = data_split(processed_full, TRAIN_START_DATE, TRAIN_END_DATE)
trade = data_split(processed_full, TRADE_START_DATE, TRADE_END_DATE)

# ---------------------------
# 4. Âª∫Á´ã Gym Áí∞Â¢É
# ---------------------------
# Determine stock dimensions and state space for the trading environment
stock_dim   = len(train.tic.unique())
state_space = 1 + 2*stock_dim + len(INDICATORS)*stock_dim

# Environment keyword arguments for the StockTradingEnv
env_kwargs = {
    'df': train,
    'hmax': 1000, # Maximum number of shares to buy/sell
    'initial_amount': 1_000_000, # Initial portfolio value
    'buy_cost_pct':    [0.001425] * stock_dim, # Buy cost percentage
    'sell_cost_pct':   [0.001425] * stock_dim, # Sell cost percentage
    'num_stock_shares':[0] * stock_dim, # Initial stock shares
    'state_space':     state_space, # Total state space dimensions
    'stock_dim':       stock_dim, # Number of unique stocks
    'tech_indicator_list': INDICATORS, # List of technical indicators used
    'action_space':    stock_dim, # Action space dimensions (number of stocks)
    'reward_scaling':  1e-4, # Reward scaling factor
    'mode': 'train' # Set mode to train for the training environment
}
e_train_gym = StockTradingEnv(**env_kwargs)

# ---------------------------
# 5. ÂÆöÁæ© SimbaV2 ÊºîÁÆóÊ≥ï
# ---------------------------
# Actor neural network for SimbaV2
class Actor(nn.Module):
    def __init__(self, s_dim, a_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(s_dim, 256), nn.ReLU(),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, a_dim), nn.Softmax(dim=-1) # Softmax for probability distribution over actions
        )
    def forward(self, x):
        return self.net(x)

# Critic neural network for SimbaV2
class Critic(nn.Module):
    def __init__(self, s_dim):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(s_dim, 256), nn.ReLU(),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Linear(128, 1) # Output a single value (V-value)
        )
    def forward(self, x):
        return self.net(x)

# SimbaV2 Reinforcement Learning algorithm
class SimbaV2:
    def __init__(self, env, gamma=0.99, actor_lr=1e-4, critic_lr=1e-3):
        self.env, self.gamma = env, gamma
        s_dim = env.observation_space.shape[0] # State dimension
        a_dim = env.action_space.shape[0] # Action dimension
        self.actor  = Actor(s_dim, a_dim) # Actor network
        self.critic = Critic(s_dim) # Critic network
        self.opt_a = optim.Adam(self.actor.parameters(), lr=actor_lr) # Optimizer for actor
        self.opt_c = optim.Adam(self.critic.parameters(), lr=critic_lr) # Optimizer for critic

    # Select an action based on the current state
    def select_action(self, s):
        with torch.no_grad():
            return self.actor(torch.FloatTensor(s).unsqueeze(0)).numpy()[0]

    # Train the SimbaV2 model for a given number of steps
    def learn(self, steps):
        obs, _ = self.env.reset() # Reset environment
        for _ in range(steps):
            action = self.select_action(obs) # Select action
            next_obs, reward, term, trunc, _ = self.env.step(action) # Take step in environment
            done = term or trunc # Check if episode is done

            s_t  = torch.FloatTensor(obs).unsqueeze(0) # Current state tensor
            s2_t = torch.FloatTensor(next_obs).unsqueeze(0) # Next state tensor
            r_t  = torch.tensor(reward, dtype=torch.float32) # Reward tensor
            # Calculate target value for critic using Bellman equation
            target = r_t + self.gamma * self.critic(s2_t).detach() * (1-int(done))

            # Critic Update: Minimize MSE loss between predicted value and target
            loss_c = nn.MSELoss()(self.critic(s_t), target)
            self.opt_c.zero_grad(); loss_c.backward(); self.opt_c.step()
            # Actor Update: Maximize expected return ( approximated by critic output * log of policy)
            loss_a = -(self.critic(s_t) * torch.log(self.actor(s_t)+1e-8)).mean()
            self.opt_a.zero_grad(); loss_a.backward(); self.opt_a.step()

            obs = next_obs if not done else self.env.reset()[0] # Update observation
        return self

    # Predict account value during trading phase
    def predict(self, env):
        env.reset() # Reset trading environment
        done = False
        while not done:
            s = env.state # Get current state
            a = self.select_action(s) # Select action
            _, _, term, trunc, _ = env.step(a) # Take step
            done = term or trunc # Check if episode is done
        return env.asset_memory  # Return recorded account values

# ---------------------------
# 6. ÈÄöÁî® DRL Agent
# ---------------------------
class DRLAgent:
    def __init__(self, env):
        self.env = env # Initialize with the Gym environment

    # Get the appropriate RL model based on name
    def get_model(self, name, **kwargs):
        if name == "simbav2":
            return SimbaV2(self.env, **kwargs)
        elif name == "a2c":
            return A2C("MlpPolicy", self.env, verbose=0, **kwargs)
        elif name == "ppo":
            return PPO("MlpPolicy", self.env, verbose=0, **kwargs)
        else:
            raise ValueError(f"Unknown model {name}")

    # Train the selected model
    def train_model(self, model, total_timesteps, name):
        if name == "simbav2":
            return model.learn(total_timesteps) # SimbaV2 has its own learn method
        else:
            model.learn(total_timesteps=total_timesteps) # Stable Baselines3 models
            return model

    @staticmethod
    # Predict account value for a given trained model and trading environment
    def predict_account(model, name, env_kwargs, trade_df):
        args = env_kwargs.copy(); args.update({'df':trade_df, 'mode':'trade'})
        e_trade = StockTradingEnv(**args) # Create a new environment for trading mode

        if name == "simbav2":
            vals = model.predict(e_trade) # SimbaV2's own predict method
        else:
            obs, _ = e_trade.reset() # Reset environment
            done = False
            vals = [e_trade.asset_memory[0]] # Initialize with initial amount
            while not done:
                action, _ = model.predict(obs) # Predict action
                obs, _, term, trunc, _ = e_trade.step(action) # Take step
                done = term or trunc # Check if episode is done
                vals.append(e_trade.asset_memory[-1]) # Append current asset value

        dates = sorted(trade_df['date'].unique())
        # If there's an extra initial value, slice it off
        if len(vals) == len(dates) + 1:
            vals = vals[1:]
        assert len(vals) == len(dates) # Ensure lengths match
        return pd.DataFrame({'date': dates, 'account_value': vals})

# ---------------------------
# 7. ÔºàÈáçË∑ëÔºâËÆ≠ÁªÉ & ÂõûÊµãÂÖ®ÈÉ®ÁÆóÊ≥ï
# ---------------------------
agent = DRLAgent(e_train_gym) # Initialize DRLAgent
algos = {"SimbaV2":"simbav2", "A2C":"a2c", "PPO":"ppo"} # Algorithms to compare
timesteps = 30 # Total timesteps for training
results = {} # Dictionary to store results
for label, name in algos.items():
    print(f"Training and evaluating {label}...")
    model   = agent.get_model(name) # Get model instance
    trained = agent.train_model(model, timesteps, name) # Train model
    df_val  = DRLAgent.predict_account(trained, name, env_kwargs, trade) # Predict account value
    results[label] = df_val # Store results

# ---------------------------
# 8. ‰∏ÄÂº†ÂõæÊØîËæÉ
# ---------------------------
plt.figure(figsize=(10,5)) # Create a figure
for label, df_val in results.items():
    plt.plot(df_val['date'], df_val['account_value'], label=label) # Plot account value for each algorithm
plt.xlabel('Date') # X-axis label
plt.ylabel('Account Value') # Y-axis label
plt.title("Comparison of Different Strategies on QCI") # Plot title
plt.legend() # Show legend
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show() # Display the plot
```
